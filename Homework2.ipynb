##### Bài 3 : Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểmtra xem nó có thể làđộ dài ba cạnh hay không, nếu không thì in ra màn hình 'Khong la ba canh cua mot tam giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra mànhình.

```python
def check1(a,b,c):
    if (a+b>c and b+c>a and c+a>b):
        return True
    else:
        return False

def check2(a,b,c):
    if (a==b or b==c or b==a):
        if (a==b and b==c):
            print ("La tam giac deu")
        else:
            if (a*a+b*b==c*c or b*b+c*c==a*a or c*c+a*a==b*b):
                print ("La tam giac vuong can")
            else:
                print ("La tam giac can")
    else:
        if (a*a+b*b==c*c or b*b+c*c==a*a or c*c+a*a==b*b):
            print ("La tam giac vuong")
        else: 
            print ("La tam giac thuong")
    
a, b, c = map(float,input().split())
if (check1(a,b,c)==False):
    print("Khong la ba canh cua mot tam giac")
else:
    check2(a,b,c)
    chuvi = a+b+c
    p = chuvi/2
    dientich = round((p*(p-a)*(p-b)*(p-c))**0.5, 2)
    print(dientich, end="cm^2 ")
    print(chuvi, end="cm")
```

##### Bài 5: Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)

```python
a, b, c = map(float,input().split())
delta = (b*b) - (4*a*c)
if (delta<0):
    print("Value Error")
elif (delta==0):
    x = (b*(-1))/(2*a)
    print (x, x)
else: 
    x1 = (b*(-1) + delta**0.5)/(2*a)
    x2 = (b*(-1) - delta**0.5)/(2*a)
    print(x1,x2)
```

##### Bài 9: Nhập vào thời gian 1 công việc nào đó là x giây. Hãy chuyển đổi và viết ra màn hình số thời gian trên dưới dạng bằng bao nhiêu giờ, bao nhiêu phút, bao nhiêu giây.

##### Bài 10: Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vàomộtđiểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?

```python
xi, yi, r = map(float,input().split())
x, y = map(float,input().split())
if (xi>x):
    dx = xi - x
else:
    dx = x - xi
if (yi>y):
    dy = yi - y
else:
    dy = y - yi
kc = (dx*dx + dy*dy)**0.5
if (kc>r):
    print ("Diem nay khong thuoc duong tron")
elif (kc==r):
    print ("Diem nay nam tren duong tron")
else:
    print ("Diem nay nam trong duong tron")
```
